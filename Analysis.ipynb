### Import Important Libraries

```python
import pandas as pd
import os
```

### Merging 12months csv data into single csv

```python
df = pd.read_csv("./Data/Sales_April_2019.csv")

files = [file for file in os.listdir('./Data')]
all_months_data = pd.DataFrame()

for file in files:
    df = pd.read_csv("./Data/"+file)
    all_months_data = pd.concat([all_months_data,df])
all_months_data.to_csv("all_data.csv",index = False)
```

Read in the updated Dataframe

```python
all_data = pd.read_csv("all_data.csv")
all_data.head()
```

### Clean up the data

#### Drop rows of NaN

```python
nan_df = all_data[all_data.isna().any(axis=1)] #axis= 1 denotes column wise

all_data = all_data.dropna(how = "all") # how = 'all' means only those rows are to be dropped whose each value is NA
```

### Drop rows with "Or"

Get rid of text in Order Date column

```python
all_data= all_data[all_data["Order Date"].str[0:2] != 'Or']
all_data.head()
```

### Convert Columns into correct type


```python
all_data['Quantity Ordered'] = pd.to_numeric(all_data['Quantity Ordered'])
all_data['Price Each'] = pd.to_numeric(all_data['Price Each'])
```

### Augment data with Additional columns

### Task 2: Add month column

```python
all_data['Month'] = all_data['Order Date'].str[0:2]
all_data['Month'] = all_data['Month'].astype('int32')
all_data.head()
```

### Task 3: Add Sales Column

```python
all_data['Sales'] = all_data['Quantity Ordered'] * all_data['Price Each']
all_data.head()
```

### Task 4: Add City Column

```python
def get_city(address):
    return address.split(',')[1]
def get_state(address):
    return address.split(',')[2].split(' ')[1]

all_data['City'] = all_data['Purchase Address'].apply(lambda x: f"{get_city(x)} ({get_state(x)})")
all_data.head()
```

### Data Exploration

### Question 1: What was the best month for sales? How much was earned that month?

```python
results = all_data.groupby('Month').sum()
```

```python
import matplotlib.pyplot as plt
months = range(1,13)

plt.bar(months, results['Sales'])
plt.xticks(months)
plt.ticklabel_format(style = 'plain')
plt.ylabel(' Sales in USD ($)')
plt.xlabel(' Month number')
plt.show()
```

### Q2. What city had the highest number of sales?


```python
city_total = all_data.groupby('City').sum()
city_total
```

```python
import matplotlib.pyplot as plt
cities = [city for city, df in all_data.groupby('City')]

plt.bar(cities, city_total['Sales'])
plt.xticks(cities, rotation= 'vertical', size=8)
plt.ylabel(' Sales in USD ($)')
plt.xlabel('City Name')
plt.show()
```

### Question #3: What time should we display advertisements to maximize the likelihood of purchases?

```python
all_data['Order Date'] = pd.to_datetime(all_data['Order Date'])
all_data.head()
```

```python
all_data['Hour'] = all_data['Order Date'].dt.hour
all_data['Minute'] = all_data['Order Date'].dt.minute
all_data.head()
```

```python
hours = [hour for hour,df in all_data.groupby('Hour')]
plt.plot(hours, all_data.groupby(['Hour']).count())
plt.xticks(hours)
plt.xlabel("Hours")
plt.ylabel("No. of orders")
plt.grid()
plt.show()


# Between 10.30 to 12am and 18 to 19Am
```

### Q4. What products are most often sold together?

```python
all_data.head()
```

```python
df = all_data[all_data['Order ID'].duplicated(keep = False)]
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))

df = df[['Order ID', 'Grouped']].drop_duplicates()
```

```python
# Referenced: https://stackoverflow.com/questions/52195887/counting-unique-pairs-of-numbers-into-a-python-dictionary

from itertools import combinations
from collections import Counter

count= Counter()

for row in df['Grouped']:
    row_list = row.split(',')
    count.update(Counter(combinations(row_list, 2)))


for key, value in count.most_common(10):
    print(key, value)
```

### Q5. What product sold the most? why do you think it sold the most?

```python
product_group = all_data.groupby('Product')
quantity_ordered = product_group.sum()['Quantity Ordered']
products = [product for product, df in product_group]


plt.bar(products, quantity_ordered)
plt.ylabel(" Quantity ordered")
plt.xlabel(" Product sold")
plt.xticks(products, rotation= 'vertical',size = 8)
plt.show()
```

```python
# Referenced: https://stackoverflow.com/questions/14762181/adding-a-y-axis-label-to-secondary-y-axis-in-matplotlib

prices = all_data.groupby('Product').mean()['Price Each']

fig, ax1 = plt.subplots()

ax2 = ax1.twinx()
ax1.bar(products, quantity_ordered, color = "green")
ax2.plot(products, prices)

ax1.set_xlabel('Product Name')
ax1.set_ylabel('Quantity Ordered', color='g')
ax2.set_ylabel('Price ($)', color='b')
ax1.set_xticklabels(products, rotation = 'vertical', size = 8)

fig.show()
```
